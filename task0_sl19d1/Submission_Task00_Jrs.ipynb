```python
def compute_mean(row):
    """
    This function computes the mean of x values in a row and asserts that it is the same as the given y.
    """
    y_calc = round(sum(row.iloc[2:])/len(row.iloc[2:]), 3)
    y_given = round(row.iloc[1], 3)
    if y_calc != y_given:
        raise ValueError(f"Missmatch claculaed {y_calc} and given {y_given} values.")
    

```
